```python
import hashlib
import random
import binascii
```

### Prueba 1

Buscamos obtener el mismo hash para el mensaje "sample", que el mostrado a continuaci√≥n.
image.png

Este hasheo se encuentra dentro del rfc-6979 en la secci√≥n A.1.2.  Generation of k. Link: https://www.rfc-editor.org/rfc/rfc6979




```python
# Se implementa el proceso de hasheo que se utiliza dentro de la generaci√≥n y verificado de firmas de nuestro proyecto
def return_hash(message):
    hash_object = hashlib.sha256()
    hash_object.update(message.encode())
    hexa = binascii.hexlify(hash_object.digest())
    return hexa
```

```python
# Prueba 1
respuesta1 = return_hash("sample")
print(respuesta1)
```

Podemos observar que el hash es el mismo en ambos casos
